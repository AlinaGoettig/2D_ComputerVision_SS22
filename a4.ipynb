### a) Was ist eine Kante und wie ist sie im Bild, in dessen Ableitung und dessen zweiten Ableitung erkennbar?
Eine Kante erkennt man an einer starken Änderung der Intensität im Bild entlang einer ausgeprägten Richtung.
In ihrer Ableitung erkennt man sie als Ausschläg in positiver oder negativer Richtung und in ihrer 2. Ableitung als Nullstellen zwischen einem hoch und einem Tiefpunkt liegt.

### b) Was sind die partiellen Ableitungen eines Bildes? Was sagen sie aus?
Die Ableitung entlang einer der Koordinatenrichtungen und zeigt Kanten an.

### c) Was ist der Gradientenvektor (kurz: Gradient) eines Bildes? Was kann aus diesem abgelesen werden?
Zeigt alle Kanten unabhängig von der Orientierung an.

### d) Wie kann die Kantenstärke berechnet werden?

### e) Wie kann die lokale Kantenrichtung berechnet werden?
Indem man die Differenz der x- & y- Werte berechnet und dessen Steigung bestimmt.

# Teil 2

```python
sobel_1 = np.array([[1,2,1]]) / 4
sobel_2 = np.array([[-1,0,1]]) / 2

def horizontal(img):
    img_copy = img.copy()
    return ut.filter(img_copy, np.transpose(sobel_1) @ sobel_2, 1, "continue")


def vertical(img):
    img_copy = img.copy()
    return ut.filter(img_copy, np.transpose(sobel_2) @ sobel_1, 1, "continue")


def gradient(img):
    return np.sqrt(horizontal(img)**2 + vertical(img)**2)


# read img
img_1 = io.imread("A4/SOBEL.png")
img_2 = io.imread("A4/dot01.png")
img_3 = io.imread("A4/dot02.png")
img_4 = io.imread("A4/fhorn.jpg")

# convert to grayscale
img_1 = ut.rgb2gray(img_1)
img_2 = ut.rgb2gray(img_2)
img_3 = ut.rgb2gray(img_3)

# plot img
ut.printImage(img_1)
ut.printImage(img_2)
ut.printImage(img_3)
ut.printImage(img_4)

# plot imgOut
imgOut_1 = horizontal(img_1)
ut.printImage(imgOut_1)

imgOut_2 = vertical(img_1)
ut.printImage(imgOut_2)

imgOut_3 = gradient(img_1)
ut.printImage(imgOut_3)

imgOut_2 = gradient(img_2)
ut.printImage(imgOut_2)
imgOut_3 = gradient(img_3)
ut.printImage(imgOut_3)
imgOut_4 = gradient(img_4)
ut.printImage(imgOut_4)
```
